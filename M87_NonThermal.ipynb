```python
from pyhamrds import bhoss
import os

fdir = '/home/astrodoo/scratch/BHOSS_EHT/bhoss_MA93/nonTH/Rh80/inc163/BHOSS_CODE'

startn = 96 
endn   = 195
ndumps = endn-startn+1

spdata = 'GRMHD_HAMR3D_EHT_MADS41A93_'+str(startn)+'_Rh80_Rl1_sig1_i163_ph0_M87_nTh_SPECTRUM.dat'
imdata = 'Rh80_Rl1_sig1_i163_ph0_M87_nTh_IMAGE_GRMHD_HAMR3D_EHT_MADS41A93_'+str(startn)+'.dat'

bhds = bhoss.read_spectrum(os.path.join(fdir,spdata), silent=False, \
                         imagefile=os.path.join(fdir,imdata))

times = np.zeros(ndumps, dtype=np.float)
nuI   = np.zeros((ndumps,len(bhds.freq)), dtype=np.float64)

for idump in range(startn, endn+1):
    spdata = 'GRMHD_HAMR3D_EHT_MADS41A93_'+str(idump)+'_Rh80_Rl1_sig1_i163_ph0_M87_nTh_SPECTRUM.dat'
    imdata = 'Rh80_Rl1_sig1_i163_ph0_M87_nTh_IMAGE_GRMHD_HAMR3D_EHT_MADS41A93_'+str(idump)+'.dat'

    bhds = bhoss.read_spectrum(os.path.join(fdir,spdata), silent=True, \
                imagefile=os.path.join(fdir,imdata))

    times[idump-startn] = bhds.time
    nuI[idump-startn]   = bhds.nuI
    
    print("%d of %d in time %f"%(idump,endn,bhds.time))
```

```python
import numpy as np
import os

pydir = os.environ['PYTHONDATA']
                  
fname = 'bhoss_specs_EHT_MADS41A93_Rh80_i163_M87_nTH.npz'
np.savez(os.path.join(pydir,fname), t=times, nuI=nuI, freq=bhds.freq, \
         th0=bhds.inclination, ph0=bhds.phi, \
         Jansky_Corr=bhds.Jansky_Corr, L_Corr=bhds.L_Corr, \
         info=[{'data':fdir,'code':'test_BHOSS', \
                'unit':{'t':'tg','nuI':'erg s^-1 cm^-2 Hz^-1 Str_px^-1', \
                        'freq':'Hz','th0':'deg','ph0':'deg'}}])
```
