# Generate M87_obs.npz files from m87_sed_2017.dat

```python
import os 
import numpy as np

def calc_xerr(f_min,f_cen,f_max):
    xerr = np.zeros((2,len(f_cen)))
    xerr[0] = f_cen - f_min
    xerr[1] = f_max - f_cen
    return xerr
    

pydir = os.environ['PYTHONDATA']
datfile = os.path.join(pydir,'m87_sed_2017.dat')

# EVN
data = np.genfromtxt(datfile, skip_header=7, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_EVN, f_cen_EVN, f_max_EVN, Fnu_EVN, err_EVN = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_EVN = calc_xerr(f_min_EVN, f_cen_EVN, f_max_EVN)

# HSA
data = np.genfromtxt(datfile, skip_header=13, max_rows=3)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_HSA, f_cen_HSA, f_max_HSA, Fnu_HSA, err_HSA = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_HSA = calc_xerr(f_min_HSA, f_cen_HSA, f_max_HSA)

# VERA
data = np.genfromtxt(datfile, skip_header=20, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_VERA, f_cen_VERA, f_max_VERA, Fnu_VERA, err_VERA = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_VERA = calc_xerr(f_min_VERA, f_cen_VERA, f_max_VERA)

# EAVN
data = np.genfromtxt(datfile, skip_header=25, max_rows=2)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_EAVN, f_cen_EAVN, f_max_EAVN, Fnu_EAVN, err_EAVN = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_EAVN = calc_xerr(f_min_EAVN, f_cen_EAVN, f_max_EAVN)

# KVN
data = np.genfromtxt(datfile, skip_header=34, max_rows=4)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_KVN, f_cen_KVN, f_max_KVN, Fnu_KVN, err_KVN = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_KVN = calc_xerr(f_min_KVN, f_cen_KVN, f_max_KVN)

# VLBA
data = np.genfromtxt(datfile, skip_header=42, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_VLBA, f_cen_VLBA, f_max_VLBA, Fnu_VLBA, err_VLBA = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_VLBA = calc_xerr(f_min_VLBA, f_cen_VLBA, f_max_VLBA)

# GMVA
data = np.genfromtxt(datfile, skip_header=49, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_GMVA, f_cen_GMVA, f_max_GMVA, Fnu_GMVA, err_GMVA = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_GMVA = calc_xerr(f_min_GMVA, f_cen_GMVA, f_max_GMVA)

# SMA
data = np.genfromtxt(datfile, skip_header=53, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_SMA, f_cen_SMA, f_max_SMA, Fnu_SMA, err_SMA = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_SMA = calc_xerr(f_min_SMA, f_cen_SMA, f_max_SMA)

# ALMA
data = np.genfromtxt(datfile, skip_header=57, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_ALMA, f_cen_ALMA, f_max_ALMA, Fnu_ALMA, err_ALMA = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_ALMA = calc_xerr(f_min_ALMA, f_cen_ALMA, f_max_ALMA)

# EHT
data = np.genfromtxt(datfile, skip_header=64, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_EHT, f_cen_EHT, f_max_EHT, Fnu_EHT, err_EHT = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_EHT = calc_xerr(f_min_EHT, f_cen_EHT, f_max_EHT)

# HST
data = np.genfromtxt(datfile, skip_header=74, max_rows=2)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_HST, f_cen_HST, f_max_HST, Fnu_HST, err_HST = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_HST = calc_xerr(f_min_HST, f_cen_HST, f_max_HST)

# Swift/UVOT
data = np.genfromtxt(datfile, skip_header=92, max_rows=6)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_Swiftu, f_cen_Swiftu, f_max_Swiftu, Fnu_Swiftu, err_Swiftu = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_Swiftu = calc_xerr(f_min_Swiftu, f_cen_Swiftu, f_max_Swiftu)

# Swift
# only the data with upper limit (5th column is the flux upper limit)
data = np.genfromtxt(datfile, skip_header=106, max_rows=1)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_Swiftx_up, f_cen_Swiftx_up, f_max_Swiftx_up, Fnu_Swiftx_up = data[:,0],data[:,1],data[:,2],data[:,4]
xerr_Swiftx_up = calc_xerr(f_min_Swiftx_up, f_cen_Swiftx_up, f_max_Swiftx_up)

# Chandra
data = np.genfromtxt(datfile, skip_header=115, max_rows=20)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_Chandra, f_cen_Chandra, f_max_Chandra, Fnu_Chandra, err_Chandra = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_Chandra = calc_xerr(f_min_Chandra, f_cen_Chandra, f_max_Chandra)

# Fermi
data = np.genfromtxt(datfile, skip_header=145, max_rows=4)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_Fermi, f_cen_Fermi, f_max_Fermi, Fnu_Fermi, err_Fermi = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
# for the upper limt data
f_min_Fermi2, f_cen_Fermi2, f_max_Fermi2, Fnu_Fermi2, err_Fermi2 = np.asarray([]),np.asarray([]),np.asarray([]),np.asarray([]),np.asarray([])
f_min_Fermi_up, f_cen_Fermi_up, f_max_Fermi_up, Fnu_Fermi_up = np.asarray([]),np.asarray([]),np.asarray([]),np.asarray([])
for i, iFnu in enumerate(Fnu_Fermi):
    if (iFnu > 1e-99):
        f_min_Fermi2 = np.append(f_min_Fermi2, f_min_Fermi[i])
        f_cen_Fermi2 = np.append(f_cen_Fermi2, f_cen_Fermi[i])
        f_max_Fermi2 = np.append(f_max_Fermi2, f_max_Fermi[i])
        Fnu_Fermi2   = np.append(Fnu_Fermi2,   Fnu_Fermi[i])
        err_Fermi2   = np.append(err_Fermi2,   err_Fermi[i])
    else:
        f_min_Fermi_up = np.append(f_min_Fermi_up, f_min_Fermi[i])
        f_cen_Fermi_up = np.append(f_cen_Fermi_up, f_cen_Fermi[i])
        f_max_Fermi_up = np.append(f_max_Fermi_up, f_max_Fermi[i])
        Fnu_Fermi_up   = np.append(Fnu_Fermi_up,   err_Fermi[i])
f_min_Fermi = f_min_Fermi2
f_cen_Fermi = f_cen_Fermi2
f_max_Fermi = f_max_Fermi2
Fnu_Fermi   = Fnu_Fermi2
err_Fermi   = err_Fermi2

xerr_Fermi = calc_xerr(f_min_Fermi, f_cen_Fermi, f_max_Fermi)
xerr_Fermi_up = calc_xerr(f_min_Fermi_up, f_cen_Fermi_up, f_max_Fermi_up)

# HESS
# only the data with upper limit (5th column is the flux upper limit)
data = np.genfromtxt(datfile, skip_header=154, max_rows=5)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_HESS_up, f_cen_HESS_up, f_max_HESS_up, Fnu_HESS_up = data[:,0],data[:,1],data[:,2],data[:,4]
xerr_HESS_up = calc_xerr(f_min_HESS_up, f_cen_HESS_up, f_max_HESS_up)

# MAGIC
data = np.genfromtxt(datfile, skip_header=164, max_rows=6)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_MAGIC, f_cen_MAGIC, f_max_MAGIC, Fnu_MAGIC, err_MAGIC = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
xerr_MAGIC = calc_xerr(f_min_MAGIC, f_cen_MAGIC, f_max_MAGIC)

# VERITAS
data = np.genfromtxt(datfile, skip_header=175, max_rows=6)
if (np.ndim(data)==1): data = data[np.newaxis,:]
f_min_VERITAS, f_cen_VERITAS, f_max_VERITAS, Fnu_VERITAS, err_VERITAS = data[:,0],data[:,1],data[:,2],data[:,3],data[:,4]
# for the upper limt data
f_min_VERITAS2, f_cen_VERITAS2, f_max_VERITAS2, Fnu_VERITAS2, err_VERITAS2 = np.asarray([]),np.asarray([]),np.asarray([]),np.asarray([]),np.asarray([])
f_min_VERITAS_up, f_cen_VERITAS_up, f_max_VERITAS_up, Fnu_VERITAS_up = np.asarray([]),np.asarray([]),np.asarray([]),np.asarray([])
for i, iFnu in enumerate(Fnu_VERITAS):
    if (iFnu > 1e-99):
        f_min_VERITAS2 = np.append(f_min_VERITAS2, f_min_VERITAS[i])
        f_cen_VERITAS2 = np.append(f_cen_VERITAS2, f_cen_VERITAS[i])
        f_max_VERITAS2 = np.append(f_max_VERITAS2, f_max_VERITAS[i])
        Fnu_VERITAS2   = np.append(Fnu_VERITAS2,  Fnu_VERITAS[i])
        err_VERITAS2   = np.append(err_VERITAS2,  err_VERITAS[i])
    else:
        f_min_VERITAS_up = np.append(f_min_VERITAS_up, f_min_VERITAS[i])
        f_cen_VERITAS_up = np.append(f_cen_VERITAS_up, f_cen_VERITAS[i])
        f_max_VERITAS_up = np.append(f_max_VERITAS_up, f_max_VERITAS[i])
        Fnu_VERITAS_up   = np.append(Fnu_VERITAS_up, err_VERITAS[i])
f_min_VERITAS = f_min_VERITAS2
f_cen_VERITAS = f_cen_VERITAS2
f_max_VERITAS = f_max_VERITAS2
Fnu_VERITAS   = Fnu_VERITAS2
err_VERITAS   = err_VERITAS2

xerr_VERITAS = calc_xerr(f_min_VERITAS, f_cen_VERITAS, f_max_VERITAS)
xerr_VERITAS_up = calc_xerr(f_min_VERITAS_up, f_cen_VERITAS_up, f_max_VERITAS_up)


f_min = np.concatenate([f_min_EVN, f_min_HSA, f_min_VERA, f_min_EAVN, f_min_KVN, f_min_VLBA, \
                        f_min_GMVA, f_min_SMA, f_min_ALMA, f_min_EHT, f_min_HST, f_min_Swiftu, \
                        f_min_Chandra, f_min_Fermi, f_min_MAGIC, f_min_VERITAS])
f_cen = np.concatenate([f_cen_EVN, f_cen_HSA, f_cen_VERA, f_cen_EAVN, f_cen_KVN, f_cen_VLBA, \
                        f_cen_GMVA, f_cen_SMA, f_cen_ALMA, f_cen_EHT, f_cen_HST, f_cen_Swiftu, \
                        f_cen_Chandra, f_cen_Fermi, f_cen_MAGIC, f_cen_VERITAS])
f_max = np.concatenate([f_max_EVN, f_max_HSA, f_max_VERA, f_max_EAVN, f_max_KVN, f_max_VLBA, \
                        f_max_GMVA, f_max_SMA, f_max_ALMA, f_max_EHT, f_max_HST, f_max_Swiftu, \
                        f_max_Chandra, f_max_Fermi, f_max_MAGIC, f_max_VERITAS])
Fnu = np.concatenate([Fnu_EVN, Fnu_HSA, Fnu_VERA, Fnu_EAVN, Fnu_KVN, Fnu_VLBA, \
                        Fnu_GMVA, Fnu_SMA, Fnu_ALMA, Fnu_EHT, Fnu_HST, Fnu_Swiftu, \
                        Fnu_Chandra, Fnu_Fermi, Fnu_MAGIC, Fnu_VERITAS])
err = np.concatenate([err_EVN, err_HSA, err_VERA, err_EAVN, err_KVN, err_VLBA, \
                        err_GMVA, err_SMA, err_ALMA, err_EHT, err_HST, err_Swiftu, \
                        err_Chandra, err_Fermi, err_MAGIC, err_VERITAS])
xerr = np.concatenate([xerr_EVN, xerr_HSA, xerr_VERA, xerr_EAVN, xerr_KVN, xerr_VLBA, \
                        xerr_GMVA, xerr_SMA, xerr_ALMA, xerr_EHT, xerr_HST, xerr_Swiftu, \
                        xerr_Chandra, xerr_Fermi, xerr_MAGIC, xerr_VERITAS],axis=1)

# upper limit data
f_min_up = np.concatenate([f_min_Swiftx_up, f_min_Fermi_up, f_min_HESS_up, f_min_VERITAS_up])
f_cen_up = np.concatenate([f_cen_Swiftx_up, f_cen_Fermi_up, f_cen_HESS_up, f_cen_VERITAS_up])
f_max_up = np.concatenate([f_max_Swiftx_up, f_max_Fermi_up, f_max_HESS_up, f_max_VERITAS_up])
Fnu_up   = np.concatenate([Fnu_Swiftx_up, Fnu_Fermi_up, Fnu_HESS_up, Fnu_VERITAS_up])
xerr_up  = np.concatenate([xerr_Swiftx_up, xerr_Fermi_up, xerr_HESS_up, xerr_VERITAS_up], axis=1)
```
