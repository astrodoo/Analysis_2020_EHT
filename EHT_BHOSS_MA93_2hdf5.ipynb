## batch script for the file conversion

```python
from pyhamrds import bhoss
import os
def bhoss2hdf5(Rhigh=1,Rlow=1,sig=1,inclination=0,phi0=0,model_name='',source='SgrA'):

    if (model_name == "MADS41A5"):
        model_str = "bhoss_MA05"
        acc = 'MAD'
        ind_start = 311
        ind_end = 410
    elif (model_name == "MADS41A0"):
        model_str = "bhoss_MA0"
        acc = 'MAD'
        ind_start = 97
        ind_end = 196
    elif (model_name == "MADS41A93"):
        model_str = "bhoss_MA93"
        acc = 'MAD'
        ind_start = 96
        ind_end = 195
    elif (model_name == "MADS41A93M"):
        model_str = "bhoss_MA93M"
        acc = 'MAD'
        ind_start = 101
        ind_end = 200
    elif (model_name == "SANEA93M"):
        model_str = "bhoss_SA93M"
        acc = 'SANE'
        ind_start=132
        ind_end = 231
    elif (model_name == "SANEA93"):
        model_str = "bhoss_SA93"
        acc = 'SANE'
        ind_start=99
        ind_end = 198
    else:
        print("no model")
        return
    
    basedir = "/home/astrodoo/projects/rrg-dhaggard/astrodoo/EHT_bhoss/BHOSS_outcomes"
    datadir = os.path.join(basedir,model_str)
    outdir = os.path.join(basedir,"bhoss2hdf5",model_str)
    
    #ind_start=400
    #ind_end=400
    for image_ind in range(ind_start,ind_end+1):
        print("%d out of %d"%(image_ind,ind_end))
        
        data_name="Rh%d_Rl%d_sig%d_i%d_ph%d_SgrA_Th_IMAGE_GRMHD_HAMR3D_EHT_%s_%d.dat" \
                  %(Rhigh,Rlow,sig,inclination,phi0,model_name,image_ind)
        
        bhdata = bhoss.read_image(os.path.join(datadir,data_name),silent=True,ifreq=0)
        bhdata.dump2hdf5(outdir=outdir,source=source,Rhigh=Rhigh,Rlow=Rlow,sigmacut=sig,acc=acc, \
                         betacrit=1., normflux=2., tilt=0., silent=True)
```
